```python
import speech_recognition as sr
import datetime
import subprocess
block_str= "0"
def time_addition(sentence,current_time):
    time_add = len(sentence.split())
    
    end_time = current_time + (datetime.timedelta(0,time_add)*.5)
    
    str_current_time = str(current_time.time())
    str_end_time= str(end_time.time())
       
                      
    with open("speech_to_text.srt", "w") as f:
                f.write(block_str)
                f.write("\n")
                f.write(str_current_time)
                f.write("-->")
                f.write(str_end_time)
                f.write("\n")
                f.write(sentence)
                f.write(".")
                f.write("\n")
                f.write("\n")
            


r = sr.Recognizer()
print("input 1 for microphone as source or input 2 for audio while as source or 3 for video as source file")
choice  = input();


if choice==1:
          
        with sr.Microphone() as source:
                print("start recording")
                audio = r.listen(source)
    
        try:
            print(r.recognize_google(audio))
        except:
            print("not found")
elif choice==2:
        print("enter the input format \n 1-flac 2-wav")
        choice1 = input()
        print("enter file name")
        name  =input()
        
        if(choice1 =="1"):
            command = "ffmpeg -i C:\\Users\\HETSHAH\\" +name + " audio3.wav"
            subprocess.call(command, shell=True)
        
        with sr.AudioFile(name) as source:
            audio1 = r.record(source)
            
        try:
            print(r.recognize_google(audio1))
        except sr.UnknownValueError:
            print("not found")
        except sr.RequestError:
            print("not found 1")
            
        print("stopped")
    
else:
    print("enter the input format \n 1.MP4 2.AVI 3.FLV ")
    choice2= input()
    print("enter file name")
    name = input()
    
    if(choice2=="1"):
        command = "ffmpeg -i C:\\Users\\HETSHAH\\" +name + " -ab 160k -ac 2 -ar 44100 -vn audio3.wav"
    
    elif(choice2=="2"):
        command = "ffmpeg -i C:\\Users\\HETSHAH\\" +name + " -ab 160k -ac 2 -ar 44100 -vn audio3.wav"
    else:
        command = "ffmpeg -i C:\\Users\\HETSHAH\\" +name + " -ab 160k -ac 2 -ar 44100 -vn audio3.wav"
    
    
    subprocess.call(command, shell=True)
    
    with sr.AudioFile("audio3.wav") as source:
        audio2 = r.record(source)
        sentence=" "
    try:
        sentence=(r.recognize_google(audio2))
    except sr.UnknownValueError:
        print("not found")
    except sr.RequestError:
        print("not found 1")
    print(sentence)
    start_time= datetime.datetime(100,1,1,0,0,0)
    time_addition(sentence,start_time)
    
        
    print("stopped")
    

            
```
